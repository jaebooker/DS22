```python
from keras.models import Sequential, Model
from keras.layers import Dense, Dropout, Activation, Lambda, Input
from keras.optimizers import SGD
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
import pandas as pd
```

```python
bdf = load_boston()
bdfs = pd.DataFrame(bdf.data)
```

```python
scaler = StandardScaler()
X = bdfs.values
y = bdf.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.1,random_state=0)
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
model = Sequential()
model.add(Dense(1,input_dim=13,activation="linear"))
model.compile(optimizer="adam",loss="mse",metrics=["mse","mae"])
model.fit(X_train, y_train, epochs=200,batch_size=1,verbose=1)
```

```python
regress = LinearRegression()
regress.fit(X_train, y_train)
```

```python
loss, mse, mae = model.evaluate(X_test,y_test, verbose=0)
sklearn_mse = mean_squared_error(regress.predict(X_test),y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```
